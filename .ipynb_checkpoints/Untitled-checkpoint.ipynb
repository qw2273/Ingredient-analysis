## Begin

Test_Pandas.xlsx contains listing information posted on a online retailer, which was sent by the company as a python test during job application. The dataset itself is messy, which can be a good practice for data cleaning. And also bring it further  .. . 

Preferred Sellers are specially selected by the retailer . They must be active sellers with high shop ratings, a large number of successful Shopee transactions, and good customer service before they can qualify to be a Shopee Preferred Seller

This dataset has 12 columns, and 464433 rows. One single listing corresponds to one row in the dataset. Here are the brief descriptions of each column:
- itemid - a unique ID of the product
- shopid - a unique ID of the shop
- item_name - product title  
- item_description - detailed  product descriptiond
- item_variation - stores variations of a product (e.g. different colours or sizes, in the format like {variation 1 name: variation 1 price, variation 2 name: variation 2 price})
- price - how much does the item sold
- stock - how many stocks left 
- category - which category does the product belongs to 
- cb_option - 1 indicates the product is sold by a cross border shop
- is_preferred - 1 indicates the product is sold by a preferred shop
- sold_count - how many products have been sold 
- item_creation_date - when are the product uploaded by the seller

### Exploratory data analysis(EDA)

```python
# a. How many unique shops are in the dataset?
len(df.shopid.unique() ) 
```

```python
#  b. How many unique preferred and cross border shops are in the dataset?
ispreferred = df.cb_option== 1
iscrossborder = df.is_preferred == 1 
len( df[ ispreferred&iscrossborder ].shopid.unique())
```

```python
# c. How many products have zero sold count?
len( df[df.sold_count==0].itemid.unique() ) 
```

```python
# d. How many products were created in the year 2018?
df.item_name.unique()
```

```python
# top five shop with largest number of unique items
by_shopid = df.groupby('shopid')
by_shopid['itemid'].count().sort_values(ascending =False).head(5)
```
